Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
rating=pd.read_csv('ratings.csv')
rating.head(5)
```

```python
def class_(column):
    if column <= 2.0:
        return 'Низкий рейтинг'
    if column <= 4.0:
        return 'Средний рейтинг'
    if column == 4.5 or column == 5.0:
        return 'Высокий рейтинг'
    return False
```

```python
rating['class_'] = rating['rating'].apply(class_)
rating.head(15)
```

Задание 2

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Результат классификации запишите в отдельный столбец region.

```python
geo=pd.read_csv('keywords.csv')
geo.head(5)
```

```python
def region(column):
    if 'москва' in column['keyword'] or 'тула' in column['keyword'] or 'ярославль' in column['keyword']:
        return 'Центр'
    if 'петербург' in column['keyword'] or 'псков' in column['keyword'] or 'мурманск' in column['keyword']:
        return 'Северо-Запад'
    if 'владивосток' in column['keyword'] or 'сахалин' in column['keyword'] or 'хабаровск' in column['keyword']:
        return 'Дальний Восток'
    
    return 'undefined'
```

```python
geo['region'] = geo.apply(region, axis=1)
geo.head(15)
```

```python
geo_centre = geo[geo['region'] == 'Центр']
geo_centre.head()
```

```python
geo_centre = geo[geo['region'] == 'Северо-Запад']
geo_centre.head()
```

```python
geo_centre = geo[geo['region'] == 'Дальний Восток']
geo_centre.head()
```
